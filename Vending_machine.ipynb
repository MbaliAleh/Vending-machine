```python
df = pd.DataFrame(list(zip(items, price_list)), 
               columns =['Items', 'Price (R)'])
df=df.set_index('Items')

print ('Welcome to the vending machine, we sell the following items:\n\n', df)

def Mbali(no_item1, no_item2, no_item3, no_item4):
    
    final_items =[]
    cost = []
    
    while (no_item1 > 0, no_item2 >0, no_item3 >0, no_item4 > 0):
        
        selected_item = input('Please choose items you would like to buy. Type N when you are finished \n')
        
        if selected_item == 'item1' or selected_item =='Item1':
            no_item1 = no_item1 - 1 
            if no_item1 < 0:
                print ('\nitem1 has been finished, please choose other items\n')
                
            else:
                print('\nYou have selected to buy item1, this item cost', item1, 'rands each\n')
                final_items.append(selected_item)
                cost.append(item1)
            
           
        elif selected_item == 'item2' or selected_item =='Item2':
            no_item2 = no_item2 - 1 
            if no_item2 < 0:
                print ('\nitem2 has been finished, please choose other items\n')
                
            else:
                print('\nYou have selected to buy item2, this item cost', item2, 'rands each\n')
                final_items.append(selected_item)
                cost.append(item2)
    
        elif selected_item == 'item3' or selected_item =='Item3':
            no_item3 = no_item3 - 1 
            if no_item3 < 0:
                print ('\nitem3 has been finished, please choose other items\n')
                
            else:
                print('\nYou have selected to buy item3, this item cost', item3, 'rands each\n')
                final_items.append(selected_item)
                cost.append(item3)
    
    
        elif selected_item == 'item4' or selected_item =='Item4':
            no_item4 = no_item4 - 1 
            if no_item4 < 0:
                print ('\nitem4 has been finished, please choose other items\n')
                
            else:
                print('\nYou have selected to buy item4, this item cost', item4, 'rands each\n')
                final_items.append(selected_item)
                cost.append(item4)
            
        elif selected_item =='N' or selected_item == 'n':
            
            df1 = pd.DataFrame(columns=['Item','Price (R)'])
            df1['Item'], df1['Price (R)'] = final_items, cost
            df2 = {'Item': 'Total_costs', 'Price (R)': sum(cost)}
            slip=df1.append(df2, ignore_index=True)
            slip=slip.set_index(['Item'])
            
            
            print ("\nHere is your bill:\n")
            print ("You have selected to purchase: ")
            print (slip)
            print ('\nPlease pay an amount of', sum(cost), 'rands.' )
            break
            
        else:
             print ("There has been an error")
                
    print ('\nThis machine only takes these coins: R1, R2, R5, R10 \nPlease insert coins!!')
    
    coins_vaule=[1,2,5,10]
    R1=int(input('\nHow many R1 coins you wish to insert, please enter only a positive interger\n'))
    R2=int(input('\nHow many R2 coins you wish to insert?, please enter only a positive interger\n'))
    R5=int(input('\nHow many R5 coins you wish to insert?, please enter only a positive interger\n'))
    R10=int(input('\nHow many R10 coins you wish to insert?, please enter only a positive interger\n'))
    
    coin_amount= coins_vaule[0]*R1 + coins_vaule[1]*R2 + coins_vaule[2]*R5 + coins_vaule[3]*R10
    print ('\nYou have inserted an amount of', coin_amount, 'rands')
    
    if coin_amount<sum(cost):
        print ('\nYou have inserted insufficient money for your selected items, your transaction cannot be processed.')
    
    elif coin_amount>=sum(cost):
        change = coin_amount - sum(cost)
        if change==0:
            print ('\nYou have no change, please take your items, THANK YOU!!')
            
        elif change>0:
            print ('\nyour change is', change, 'rands, please take your change and your items, THANK YOU!!')
    else:
        print ("There has been an error")
        
Mbali(2,2,2,2)           
```
